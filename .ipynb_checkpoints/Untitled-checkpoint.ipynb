```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, classification_report, accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
train = pd.read_csv("/Users/navamongkoltongta/.kaggle/Titanic/train_data.csv")
test = pd.read_csv("/Users/navamongkoltongta/.kaggle/Titanic/test_data.csv")
```

```python
train.Fare.value_counts()
```

```python
train['ratio'] = train['Family_size']/train['Fare']
```

```python
train.dropna(inplace=True)
```

```python
# X = train[['Sex', 'Age', 'Fare',
#        'Pclass_1', 'Pclass_2', 'Pclass_3', 'Family_size', 'Title_1', 'Title_2',
#        'Title_3', 'Title_4', 'Emb_1', 'Emb_2', 'Emb_3']]
# X = train[['Sex','Age','Title_1','Pclass_3','Emb_3']]
X = train[['Sex', 'Age','Family_size','Fare','Pclass_1', 'Pclass_2', 'Pclass_3','Title_1', 'Title_2',
       'Title_3', 'Title_4', 'Emb_1', 'Emb_2', 'Emb_3','ratio']]
y = train[['Survived']]
```

```python
steps = [('rf',RandomForestClassifier())]
```

```python
pipeline = Pipeline(steps)
```

```python
parameters = {'rf__n_estimators': [5,10,15,20],
              'rf__max_features': ['auto','sqrt','log2'],
             'rf__max_depth':[50,100,150],
             'rf__min_samples_leaf':[1,2,3,4,5],
             'rf__min_samples_split':[1,2,3,4,5]}
grid = GridSearchCV(pipeline, param_grid=parameters, cv=5)
```

```python
# corr = train.corr()
# sns.set(rc={'figure.figsize':(11,8)})
# sns.heatmap(round(corr,2), annot=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
grid.fit(X_train, y_train.values.ravel())
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
train_score = grid.score(X_train, y_train)
test_score = grid.score(X_test, y_test)
print(f"train : {train_score}")
print(f"test : {test_score}")
```

```python
cm = confusion_matrix(y_pred, y_test)
```

```python
sns.heatmap(cm, annot = True , cbar=False)
```

```python
#Predict Accuracy
acc = accuracy_score(y_test, y_pred)
error = np.sqrt(mean_squared_error(y_test, y_pred))
overfitting_by = grid.score(X_train, y_train)*100 - grid.score(X_test, y_test)*100
print("Accuracy :",acc)
print("Error :",error)
print("Overfitting :", overfitting_by)
```
